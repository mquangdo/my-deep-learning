```python
import osmnx as ox
import matplotlib.pyplot as plt

place_name = "Nam Từ Liêm, Hanoi, Vietnam"


hanoi = ox.geocode_to_gdf(place_name)

fig, ax = plt.subplots(figsize=(8, 8))
hanoi.plot(ax=ax, color="lightblue", edgecolor="black")

plt.title('Hà Đông từ OpenStreetMap')
plt.xlabel('Kinh độ')
plt.ylabel('Vĩ độ')
plt.show()
```

```python
# Tải dữ liệu mạng lưới đường phố của Hà Nội
G = ox.graph_from_place(place_name, network_type='drive')

# Vẽ mạng lưới đường phố lên bản đồ
fig, ax = plt.subplots(figsize=(8, 8))
ox.plot_graph(ox.project_graph(G), ax=ax, node_size=0, edge_color='black', edge_linewidth=0.5)
```
