```python
import osmnx as ox
import matplotlib.pyplot as plt

place_name = "Nam T Lim, Hanoi, Vietnam"


hanoi = ox.geocode_to_gdf(place_name)

fig, ax = plt.subplots(figsize=(8, 8))
hanoi.plot(ax=ax, color="lightblue", edgecolor="black")

plt.title('H ng t OpenStreetMap')
plt.xlabel('Kinh ')
plt.ylabel('V ')
plt.show()
```

```python
# Ti d liu mng li ng ph ca H Ni
G = ox.graph_from_place(place_name, network_type='drive')

# V mng li ng ph ln bn 
fig, ax = plt.subplots(figsize=(8, 8))
ox.plot_graph(ox.project_graph(G), ax=ax, node_size=0, edge_color='black', edge_linewidth=0.5)
```
